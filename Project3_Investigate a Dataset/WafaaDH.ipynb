```python
df.dropna(inplace=True)
df.info()
```

```python
#1: What is the name and type of the most popularity movi?
most_popu=df['popularity'].max()
most_popu
```

```python
most_popu_name=df['original_title'].loc[df['popularity']==most_popu]
most_popu_genres=df['genres'].loc[df['popularity']==most_popu]
most_popu_name,most_popu_genres
```

```python
#2: What movie that have highest revenues?
higest_revenu_movi=df['revenue'].max()
higest_revenu_movi
```

```python
name_higest_revenu_movi=df['original_title'].loc[df['revenue']== higest_revenu_movi]
name_higest_revenu_movi
```

```python
# lets see the describtion of Avatar movi
df.loc[df['original_title']=='Avatar'].describe()
```

```python
#3: What is the relationship between budjet spend on the movi and How many it runs ?
plt.plot(df['release_year'],df['runtime'] )
plt.legend();
plt.xlabel("release year")
plt.ylabel("number of runtime")
plt.title("relationship between budjet spend and num of runs")
plt.show()
#The chart is confusing because I think there is more than one 
#movie released in the same year, and it is normal that each film has a number of different views.
```

```python
#4: lets compere between the number of Horror & Comedy movi
df.loc[df['genres']=='Horror'].count().plot(kind='bar', alpha=0.5, color='blue' ,label='Horror')
df.loc[df['genres']=='Comedy'].count().plot(kind='bar', alpha=0.5, color='red',label='Comedy')
plt.legend();
plt.ylabel("number of movi")
plt.title("Horror Vs Comedy Movi")
plt.show()
#It is clear from this chart that people tend to COMEDY movies more than HORROR movies so the directors resort to making comedies one, because the painting indicates that the number of comedy movies more than horror movies
```
